```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
lenguajes_de_programación = [ " Python", "JavaScript", "C++", "C##", "Java"]
print(lenguajes_de_programación)
```

```python
preferencia = [45, 20, 15, 12, 8]
print(preferencia)

plt.figure(figsize = (8,4))
plt.bar(lenguajes_de_programación,preferencia, color = ['Green','Blue', 'Red', 'Yellow', 'Black'])
plt.title('Preferencia de Lenguajes de Programación 2026', fontsize=12, fontweight='bold')
plt.xlabel('Lenguajes de Programación', fontsize=12)
plt.ylabel('Porcentaje de Preferencia (%)', fontsize=12)

plt.show()
```

```python
meses = [ "Enero", "Febrero", "Marzo", "Abril" "Mayo"]
print(meses)
```

```python
ventas = [100, 110, 80, 40, 55]
print(ventas)

plt.figure(figsize=(8,4))
plt.bar(meses, ventas, color=["red","yellow","darkgreen","blue","indigo"])
plt.title("Ventas durante 5 meses", fontsize=12, fontweight="bold")
plt.xlabel("Meses", fontsize=12)
plt.ylabel("Ventas", fontsize=12)

plt.show()
```
