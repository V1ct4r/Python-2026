***Ejercicio Practica***

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
facultades = ["Ing", "Cs. de la educación", "Cs. sociales", "Salud", "Artes"]
proyectos = [8, 4, 19, 16, 3]
print(facultades)
print(proyectos)
```

```python
plt.figure(figsize=(8,4))
plt.bar(facultades, proyectos, color =["Indigo", "Blue", "Red", "Yellow", "Green"])

plt.title('Proyectos de Investigación - Universidad de Los Lagos', fontsize = 12, fontweight = 'bold')
plt.xlabel('Facultades', fontsize = 12 )
plt.ylabel('Proyectos', fontsize = 12 )

plt.show()
```
