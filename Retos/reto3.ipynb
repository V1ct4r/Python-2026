```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def preparar_datos():
    horas_simulacion = np.arange(1,9) # array = [1,2,3,4,5,6,7,8]
    gpu =[12.5, 14.2, 18.1, 22.8, 25.4, 21.0, 17.5, 15.3]
    
    data ={
        'Horas': horas_simulacion,
        'Uso': gpu
    }

    # Transofmrando el diccionario a un data frame
    df = pd.DataFrame(data)
    return df
```

```python
def graficar_rendimiento(dataframe):
    fig, ax = plt.subplots() # -> ax = axes (ejes / fig = figure)

    ax.plot(dataframe['Horas'], dataframe['Uso'], marker='o', color='Yellow', linestyle='-')
    ax.set_title(' Monitoreo de la infraestructura tecnológica')
    ax.set_xlabel('Tiempo (Horas)')
    ax.set_label('Caudal (m³/s)')

    ax.grid(True)


    
```

```python
gpu = preparar_datos()# invocando la primera funcion 
graficar_rendimiento(gpu) # invocamdo segunda funcion con el primer parametro de la funcion
plt.show()
```
